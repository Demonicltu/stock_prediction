```python
def execute_benchmark_full_program_for_specific_financial_unit(unit):
  download_data(unit)

  real_prices, predicted_prices = get_benchmark_predicted_prices()

  plot_graphic(real_prices, predicted_prices, unit + '-BENCHMARK')
    
  mape_error = calculate_MAPE_error(real_prices[['Price']], predicted_prices[['Price']])
  mse_error = calculate_MSE_error(real_prices[['Price']], predicted_prices[['Price']])
  
  print()
  print("MAPE: %3.3f %%" % (mape_error))
  print("MSE: ", (mse_error))

def get_benchmark_predicted_prices():
  dataset = pd.read_csv("data.csv")
  dataset = dataset.dropna(how='any',axis=0)

  real_prices = dataset[['Date']]
  real_prices = real_prices.assign(Price = dataset['Open'])

  dataset['Price'] = dataset['Open'].shift(1)

  dataset.at[0, 'Price'] = dataset.at[0, 'Open']

  predicted_prices = dataset[['Date', 'Price']]

  predicted_prices['Date'] = pd.to_datetime(predicted_prices['Date'])
  real_prices['Date'] = pd.to_datetime(real_prices['Date'])

  return real_prices, predicted_prices

# MAPE - vidutinė procentinė absoliutinė paklaida
def calculate_MAPE_error(Y_actual,Y_Predicted):
    return np.mean(np.abs((Y_actual - Y_Predicted) / Y_actual)) * 100

# MSE - Vidutinė kvadratinė prognozės paklaida
def calculate_MSE_error(Y_actual,Y_Predicted):
    return np.square(np.subtract(Y_actual, Y_Predicted)).mean()
```

```python
#@title
def execute_full_program_for_specific_financial_unit(unit):
  download_data(unit)

  sc = MinMaxScaler(feature_range=(0,1))

  dataset_train, dataset_test = get_train_test_datasets()

  X_train, y_train = get_train_data_structures(dataset_train, sc)

  model = get_rnn_model(X_train)

  model = train_model(model, X_train, y_train)

  real_prices, predicted_prices = get_real_predicted_prices(dataset_train, dataset_test, model, sc)

  plot_graphic(real_prices, predicted_prices, unit)

  mape_error = calculate_MAPE_error(real_prices[['Price']], predicted_prices[['Price']])
  mse_error = calculate_MSE_error(real_prices[['Price']], predicted_prices[['Price']])
  
  print()
  print("MAPE: %3.3f %%" % (mape_error))
  print("MSE: ", (mse_error))
  print()
  print('Benchmark modeliu prognozuota')
  print()

  execute_benchmark_full_program_for_specific_financial_unit(unit)
```

```python
#@title
#--------------------- Data Preprocessing --------------------#
# Importing and scaling the data
  
def get_train_test_datasets(test_size = 0.2):  
  dataset = pd.read_csv("data.csv")

  dataset = dataset.dropna(how='any',axis=0) 

  num_train = int((1-test_size)*len(dataset))

  dataset_train = dataset[:num_train]
  dataset_test = dataset[num_train:]
  return dataset_train, dataset_test

def get_train_data_structures(dataset_train, sc):
  #selecting the right column (we need all rows and column 1) : numpy array
  training_set = dataset_train.iloc[:,1:2].values

  # Feature scaling
  training_set_scaled = sc.fit_transform(training_set)

  #creating a data structure with 60 timesteps and 1 output
  X_train = []
  y_train = []

  for i in range(60,training_set_scaled.size):
      # appending the 60 previous stock prices to the list for i
      # we need to specify the rows and simply pick the first and only column
      X_train.append(training_set_scaled[i-60:i, 0])
      # appending the 60th stock price to the list for i
      y_train.append(training_set_scaled[i, 0])
  # transforming pandas lists to numpy arrays required for the RNN
  X_train, y_train = np.array(X_train), np.array(y_train)
  
  # Shaping/adding new dimensions to allow adding more indicators: from 2D to 3D
  # 3 input arguments
  # batch_size: number of observations
  # timesteps: number of columns
  # input_dim: number of predictors
  X_train = np.reshape(X_train, [X_train.shape[0], X_train.shape[1], 1])
  
  return X_train, y_train

#--------------------- Building RNN/LSTM model --------------------#
#Initializing the RNN
def get_rnn_model(X_train):
  # independent variable is a sequence of layers: regression and no classification given the continuous output value
  regressor = Sequential()
  #------------------------------------------------------------------#
  # Adding the first LSTM layer
  #------------------------------------------------------------------#

  # 3 inputs
  # number of memory/LSTM units or neurons in each LSTM
  # binary vb to indicate whether there will be further layers of LSTM added to teh model
  # input shape (automatically takes teh first dimension so the reamining only needs to be specified)
  regressor.add(LSTM(units = 50, return_sequences=True, input_shape = (X_train.shape[1], 1)))

  # adding Dropout regularization layers
  # 1 input: amount of neurons to ignore in the layers
  regressor.add(Dropout(0.2))

  #------------------------------------------------------------------#
  # Adding the second LSTM layer
  #------------------------------------------------------------------#
  # no input shape needed given that that is specified in the previous layer
  regressor.add(LSTM(units = 50, return_sequences=True))
  regressor.add(Dropout(0.2))

  #------------------------------------------------------------------#
  # Adding the third LSTM layer
  #------------------------------------------------------------------#
  regressor.add(LSTM(units = 50, return_sequences=True))
  regressor.add(Dropout(0.2))

  #------------------------------------------------------------------#
  # Adding the forth LSTM layer
  #------------------------------------------------------------------#
  regressor.add(LSTM(units = 50, return_sequences=True))
  regressor.add(Dropout(0.2))

  #------------------------------------------------------------------#
  # Adding the fifth LSTM layer
  #------------------------------------------------------------------#
  # note that this is the final LSTM layer, hence we change the binary argument to False
  regressor.add(LSTM(units = 50))
  regressor.add(Dropout(0.2))

  #------------------------------------------------------------------#
  # Adding output layer to the RNN to make a fully connected NN
  #------------------------------------------------------------------#
  # one dimensional real output
  regressor.add(Dense(units = 1))


  #--------------------- Compiling the RNN model --------------------#
  regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')

  return regressor

#--------------------- Training RNN model --------------------#
#connecting the built regressor to the training model
def train_model(model, X_train, y_train, epochs = 120, batch_size = 32):
  model.fit(X_train, y_train, epochs = epochs, batch_size = batch_size)

  return model

# dataset_train, dataset_test = get_train_test_datasets()
# X_train, y_train = get_train_data_structures(dataset_train)
# model = get_rnn_model(X_train)
# model = train_model(model, X_train, y_train)
```

```python
#@title
#--------------------- Testing RNN model --------------------#
def get_real_predicted_prices(dataset_train, dataset_test, regressor, sc):
  # actual stock prices
  real_stock_price = dataset_test.iloc[:,1:2].values
  # predicting the stock prices using X_test
  # we need the original training data (vertical concatination 0, horizontal 1)
  dataset_total = pd.concat((dataset_train['Open'], dataset_test['Open']), axis = 0)
  # print(real_stock_price)

  # for the consistence of the model we need to have the same scaling on the test as on the training
  # for the first test day we need the previous 60 days data from train
  inputs = dataset_total[len(dataset_total) - len(dataset_test) - 60:].values

  # reshaping,normalizing the inputs
  inputs = inputs.reshape(-1,1)
  # no longer fitting but just transforming
  inputs = sc.transform(inputs)

  #storing test data as we did in teh training
  X_test = []
  for i in range(60,inputs.size):
      X_test.append(inputs[i-60:i, 0])
  X_test = np.array(X_test)
  #adding extra dimension
  X_test = np.reshape(X_test, [X_test.shape[0], X_test.shape[1], 1])

  #obtaining predicted values
  predicted_stock_price = regressor.predict(X_test)
  predicted_stock_price = sc.inverse_transform(predicted_stock_price)

  price_df = pd.DataFrame(predicted_stock_price, columns=['Price'])

  predicted_prices = dataset_test[['Date']]
  predicted_prices['Date'] = pd.to_datetime(predicted_prices['Date'])
  predicted_prices.insert(1, 'Price', price_df['Price'].values)

  real_prices = dataset_test[['Date']]
  real_prices['Date'] = pd.to_datetime(real_prices['Date'])
  real_prices.insert(1, 'Price', dataset_test['Open'])

  return real_prices, predicted_prices
```

```python
#@title
#--------------------- Visualizing the RNN model results--------------------#
def plot_graphic(real_prices, predicted_prices, diagram_name):
  plt.figure(figsize=(30, 15))
  plt.plot(real_prices['Date'], real_prices['Price'], color = 'blue', label = "Tikra akcijų kaina")
  plt.plot(predicted_prices['Date'], predicted_prices['Price'], color = 'red', label = "Prognozuota akcijų kaina")
  plt.title("Akcijų kainų laiko eilučių prognozė")

  plt.xlabel("Data")
  plt.ylabel("Kaina")
  plt.legend()
  plt.grid()
  plt.savefig(diagram_name + '.png')
  plt.show()
```
